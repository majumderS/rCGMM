```python
def centre_of_mass(bead):
    x = 0.0
    y = 0.0
    z = 0.0
    
    l= len(bead['x'])
    tot_mass=0.0
    for m in bead['mass']:
        tot_mass = tot_mass+m
        
    for i in range(0,l):
        x = x+(float(bead['x'][i])*bead['mass'][i])
        
    for i in range(0,l):
        y = y+(float(bead['y'][i])*bead['mass'][i])
        
    for i in range(0,l):
        z = z+(float(bead['z'][i])*bead['mass'][i])
        

#     print(x,y,z)
#     print(tot_mass)
    x = x/tot_mass
    y = y/tot_mass
    z = z/tot_mass
#     print(x,y,z)
    return round(x,2),round(y,2),round(z,2)
        
```
