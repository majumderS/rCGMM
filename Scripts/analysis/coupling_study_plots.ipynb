```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax1 = plt.subplots()
fig.set_figheight(10)
fig.set_figwidth(10)
color = 'tab:red'
ax1.set_xlabel('tau (ps)')
ax1.set_ylabel(u'Δ energy', color=color)
ax1.set_ylim(200,1600)
ax1.plot(tau, delta_energy_vrescale_sys1, color="red",label=u'Δ energy system level coupling')
ax1.plot(tau, delta_energy_vrescale_sys2, color="blue",label=u'Δ energy separate uniform coupling')
ax1.plot(tau, delta_energy_vrescale_sys3, color="green",label=u'Δ energy separate non-uniform coupling')

ax1.tick_params(axis='y', labelcolor=color)


ax2 = ax1.twinx()  # instantiate a second Axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel(u'Δ temp', color=color)  # we already handled the x-label with ax1
ax2.plot(tau, delta_temp_vrescale_sys1, color="red",label=u'Δ temp system level coupling',ls="dashdot")
ax2.plot(tau, delta_temp_vrescale_sys2, color="blue",label=u'Δ temp separate uniform coupling',ls="dashdot")
ax2.plot(tau, delta_temp_vrescale_sys3, color="green",label=u'Δ temp separate non-uniform coupling',ls="dashdot")
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylim(0,100)
# plt.legend()

# ask matplotlib for the plotted objects and their labels
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2,loc='lower center', bbox_to_anchor=(0.5, -0.25),
          ncol=2, fancybox=True, shadow=True)

plt.title("V-rescale Energy fluctuations over tau")

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
fig.savefig("energy_fluctuation_all_system_V-rescale.png")
```
