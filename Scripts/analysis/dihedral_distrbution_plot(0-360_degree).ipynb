```python
# Density Plot and Histogram 
l = []
for item in BB1_BB2_BB3_pBB1_Hmapped_dihedral.values:
    l.append(int(item+180))
    
l = pd.Series(l)
plt = sns.distplot(l, kde=True, 
            color = 'darkblue', norm_hist=True,
             kde_kws={'linewidth': 4})
# plt.set_xlabel('BB1_BB2_BB3_ASC1(nm)'+'\nMean: '+str("{:.2f}".format(l.mean()))+",\
# SD: "+str("{:.2f}".format(l.std())))
# plt.set(xlim=(0,360))


data1 = []
data2 = []
for item in l.values:
    if(item>=0 and item <200 ):
        data1.append(item)
    elif(item>=200 and item <360):
        data2.append(item)

data1 = pd.Series(data1)
data2 = pd.Series(data2)


plt.set_xlabel('BB1_BB2_BB3_pBB1_Hmapped_dihedral(nm)'+'\n\n(Peak1) Mean: '+str("{:.2f}".format(data1.mean()))+", \
SD: "+str("{:.2f}".format(data1.std())) + '\n\n(Peak2) Mean: '+str("{:.2f}".format(data2.mean()))+", \
SD: "+str("{:.2f}".format(data2.std())))


dihedral_dict['BB1_BB2_BB3_pBB1'] = [str("{:.2f}".format(data1.mean())),str("{:.2f}".format(data2.mean())),str("{:.2f}".format(data1.std())),str("{:.2f}".format(data2.std()))]
plt.figure.savefig("BB1_BB2_BB3_pBB1.png")
```
