```python
``

# Access the y-values of the KDE
data1 = np.random.randn(100)
data2 = np.random.randn(150) + 2

# Plot multiple KDE plots with y-values bound between 0 and 1
plt.figure(figsize=(10, 6))

# KDE for data1
kde1 = sns.kdeplot(data1, fill=True, label='Data 1')
kde_values1 = kde1.get_lines()[0].get_ydata()
normalized_kde_values1 = kde_values1 / np.trapz(kde_values1, kde1.get_lines()[0].get_xdata())

# KDE for data2
kde2 = sns.kdeplot(data2, fill=True, label='Data 2')
kde_values2 = kde2.get_lines()[0].get_ydata()
normalized_kde_values2 = kde_values2 / np.trapz(kde_values2, kde2.get_lines()[0].get_xdata())

# Plot the original and normalized KDEs
plt.plot(kde1.get_lines()[0].get_xdata(), normalized_kde_values1, label='Normalized KDE (Data 1)')
plt.plot(kde2.get_lines()[0].get_xdata(), normalized_kde_values2, label='Normalized KDE (Data 2)')

plt.legend()
plt.title('Multiple KDE Plots with Y-Values Bound 0-1')
plt.xlabel('X')
plt.ylabel('Normalized KDE Value')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the standard normal PDF function
def standard_normal_pdf(x):
    return 1 / np.sqrt(2 * np.pi) * np.exp(-0.5 * x**2)

# Generate a range of x values
x_values = np.linspace(-3, 3, 1000)

# Calculate the PDF values for each x
pdf_values = standard_normal_pdf(x_values)

# Plot the PDF
plt.plot(x_values, pdf_values, label='Standard Normal PDF')
plt.title('Probability Density Function (PDF)')
plt.xlabel('X')
plt.ylabel('PDF Value')
plt.legend()
plt.show()
```
