```python
import os
import math
```

```python

#contains water,lipids,sugar,heme,metals and ions
#taken from http://prody.csb.pitt.edu/manual/reference/atomic/flags.html  https://proteopedia.org/wiki/index.php/Hetero_atoms
non_standard_residue = ['HOH','DOD','AL','AL','BA','CA','CD','CL','CO','CS','CU','CU1','CUA','HG','IN','IOD','K',
                        'MG','MN3','NA','PB','PT','RB','TB','TL','WO4','YB','ZN','CAL','CES','CLA','POT','SOD','ZN2',
                       'SO4','PO4','SR','CBV','A23','IR','GPE','LPP','OLA','SDS','STE','POPC','LPPC','POPE','DLPE',
                        'PCGL','STEA','PALM','OLEO','DMPC','BGC','GLC','GLO','AGLC','AMP','ADP','CDP',
                        'GMP','GDP','TMP','UMP','UDP','1FH','2FH','DDH','DHE','HAS','HDD','HDE',
                        'HDM','HEA','HEB','HEC','HEM','HEO','HES','HEV','NTE','SRM','VER','HEMO','HEMR','CSL','UMS',
                        'CBR','BA','5BU','CCC','GLC','LI','AL','SI','CR','CA','SC','TI','V','MN','FE','CO','NI','CU',
                        'Zn','GA','GE','AS','SE','BR','KR','RB','SR','Y','ZR','NB','MO','TC','RU','RH','PD','AG','CD',
                        'IN','SN','SB','TE','I','XE','CS','BA','LA','CE','PR','ND','PM','SM','EU','GD','TB','DY','HO',
                        'ER','TM','YB','LU','HF','TA','W','RE','OS','IR','PT','AU','HG','TL','PB','BI','PO','AT','RN',
                        'FR','RA','AC','TH','PA','U','NP','PU','AM','CM','BK','CF','ES','FM','MD','NO','LR','RF','DB',
                        'SG','BH','HS','MT','DS','RG','CN','NH','FL','MC','LV','TS','OG']

ligand = ['ARG','SAM','QSA','SS0','AM2','EKJ','J8U','DSH','MTA','ADN','RIO','ARG','ISI','AM2','MGR','SRY','SPM',
         'FMN','BTN','QSY','PAR','ISH','JS5','TEL','ROS','QSV','HMV','QW4','HLV','EKM','UYS','HMJ','FRU','38E','ARG',
         'PRF','JS4','GTP','UTP','TTP','ATP','CTP','TEP','PAR','LIV']
set(ligand).intersection(set(non_standard_residue))
```

```python
data = {}
hetatms = []
available_ligands = []

for file in os.listdir("../Dataset/Hariboss_RNA_Dataset/"):
    print(file)
    target_coords = []
    ligand_coords = []
    BS_ids = []
    NBS_ids = []

    with open ("../Dataset/Hariboss_RNA_Dataset/"+file) as f:
        for line in f.readlines():

            #condition to pick first model, terminate on raching eof
            if(line.split()[0] == 'ENDMDL' or line.split()[0] == 'END'):
                break

            #collect coordinates in a list for atoms and hetatoms
            elif (line.split()[0] == 'ATOM' or line.split()[0] == 'HETATM'):
#                 print(line)
                if(line.split()[0] == 'ATOM'):
                    target_coords.append([line.split()[1],line.split()[5],line.split()[6],line.split()[7],line.split()[8]])
                elif(line.split()[0] == 'HETATM' and line.split()[3] not in non_standard_residue): 
                    #flag hetero non-ligand
#                     print(line)
                    available_ligands.append((file,line.split()[3]))
                    ligand_coords.append([line.split()[1],line.split()[5],line.split()[6],line.split()[7],line.split()[8]])

        #flag target_atom as BS or NBS based on euclidean distance threshold of 4 Angstrom
        for atom_i in target_coords:
            near_ligand = False
            for atom_j in ligand_coords:
                xx = float(atom_i[2]) - float(atom_j[2])
                yy = float(atom_i[3]) - float(atom_j[3])
                zz = float(atom_i[4]) - float(atom_j[4])
                eu_dis = math.sqrt(xx*xx + yy*yy + zz*zz)
                if(eu_dis < 5.0):
                    near_ligand = True
                    break

            if(near_ligand):
                BS_ids.append((atom_i[0],atom_i[1]))
            else:
                NBS_ids.append((atom_i[0],atom_i[1]))
    data[file]={'NBS':NBS_ids,'BS':BS_ids}
```
